```python
#Delete the team names which are not playing
teams = np.delete(teams, np.where(teams == 'Rising Pune Supergiants')[0])
teams = np.delete(teams, np.where(teams == 'Kochi Tuskers Kerala')[0])
teams = np.delete(teams, np.where(teams == 'Rising Pune Supergiant')[0])
teams = np.delete(teams, np.where(teams == 'Gujarat Lions')[0])
teams = np.delete(teams, np.where(teams == 'Pune Warriors')[0])
teams = np.delete(teams, np.where(teams == 'Deccan Chargers')[0])
```

```python
#Changing the team names
match_df['team1'] = match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2'] = match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['team1'] = match_df['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['team2'] = match_df['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['team1'] = match_df['team1'].str.replace('Royal Challengers Bangalore','Royal Challengers Bangaluru')
match_df['team2'] = match_df['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bangaluru')

match_df['team1'] = match_df['team1'].str.replace('Kings XI Punjab','Panjab kings')
match_df['team2'] = match_df['team2'].str.replace('Kings XI Punjab','Panjab kings')

match_df['toss_winner'] = match_df['toss_winner'].str.replace('Royal Challengers Bangalore','Royal Challengers Bangaluru')
match_df['toss_winner'] = match_df['toss_winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['toss_winner'] = match_df['toss_winner'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['toss_winner'] = match_df['toss_winner'].str.replace('Kings XI Punjab','Panjab kings')

match_df['winner'] = match_df['winner'].str.replace('Kings XI Punjab','Panjab kings')
match_df['winner'] = match_df['winner'].str.replace('Royal Challengers Bangalore','Royal Challengers Bangaluru')
match_df['winner'] = match_df['winner'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['winner'] = match_df['winner'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
match_df = match_df[match_df['team1'].isin(teams)]
match_df = match_df[match_df['team2'].isin(teams)]
match_df = match_df[match_df['toss_winner'].isin(teams)]
match_df = match_df[match_df['winner'].isin(teams)]
```

```python
match_df
```

```python
# DL applied matches are removed
match_df = match_df[match_df['dl_applied'] == 0]
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
delivery_df = match_df.merge(df_delivery,on='match_id')
```

```python
delivery_df = delivery_df[delivery_df['inning']==2]
```

```python
delivery_df.shape
```

```python
delivery_df
```

```python
# we need to know how much run scorred after each ball (cumlativeSum)
delivery_df['current_score'] = delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
#run left
delivery_df['runs_left'] = delivery_df['total_runs_x'] - delivery_df['current_score']
delivery_df['runs_left'] = delivery_df['runs_left'].apply(lambda x: max(x, 0))
```

```python
delivery_df
```

```python
# Balls left after each delivery
delivery_df['balls_left'] = 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
#Wickets left
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0") # fill in missing values with '0'
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x: 1 if str(x) != "0" else 0) # convert non-zero strings to 1
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int') # convert 'player_dismissed' column to integer type
```

```python
wickets = delivery_df.groupby('match_id')['player_dismissed'].transform('cumsum')  # calculate cumulative sum of wickets for each match
delivery_df['wickets'] = 10 - wickets # calculate remaining wickets
```

```python
delivery_df
```

```python
# crr = runs/overs
delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
delivery_df['crr']
```

```python
#RRR = runs/balls left
delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0

delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
#suffle the Records
final_df = final_df.sample(final_df.shape[0])
final_df.dropna(inplace=True)
final_df = final_df[final_df['balls_left'] != 0]
```

```python
from sklearn.model_selection import train_test_split
X = final_df.iloc[:,:-1]# all rowsand column accept last
y = final_df.iloc[:,-1]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
#This allows you to apply different preprocessing techniques to different columns of your data,
# making your data ready for machine learning models.
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
#Pipeline
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
def match_summary(row):
    print("Batting Team-" + row['batting_team'] + " | Bowling Team-" + row['bowling_team'] + " | Target- " + str(row['total_runs_x']))
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
def match_progression(x_df,match_id,pipe):
  # Filter for the specific match
    match = x_df[x_df['match_id'] == match_id]
  # Filter for the end of each over
    match = match[(match['ball'] == 6)]
  # select relevant column from five dataset
    temp_df = match[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr']].dropna()
  # filter where inning is ended
    temp_df = temp_df[temp_df['balls_left'] != 0]
  # predict the probability
    result = pipe.predict_proba(temp_df)
    temp_df['lose'] = np.round(result.T[0]*100,1)
    temp_df['win'] = np.round(result.T[1]*100,1)
    temp_df['end_of_over'] = range(1,temp_df.shape[0]+1)
  # After each ball values update
    target = temp_df['total_runs_x'].values[0]
    runs = list(temp_df['runs_left'].values)
    new_runs = runs[:]
    runs.insert(0,target)
    temp_df['runs_after_over'] = np.array(runs)[:-1] - np.array(new_runs)
    wickets = list(temp_df['wickets'].values)
    new_wickets = wickets[:]
    new_wickets.insert(0,10)
    wickets.append(0)
    w = np.array(wickets)
    nw = np.array(new_wickets)
    temp_df['wickets_in_over'] = (nw - w)[0:temp_df.shape[0]]
  # Print
    print("Target-",target)
    temp_df = temp_df[['end_of_over','runs_after_over','wickets_in_over','lose','win']]
    return temp_df,target
```

```python
temp_df,target = match_progression(delivery_df,11318,pipe)
temp_df
```
